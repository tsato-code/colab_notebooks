```python
import datetime
import os
import subprocess
```

```python
# 保存先のディレクトリ
out_dir = '/content/drive/MyDrive/works/data/uma/'
os.makedirs(out_dir, exist_ok=True)
```

```python
# データ取得開始日と終了日を指定
start = datetime.datetime.strptime('2021-05-01', '%Y-%m-%d')
end   = datetime.datetime.strptime('2021-05-02', '%Y-%m-%d')

# データ取得日のリストを作成
dt_list = []
for _i in range((end - start).days):
    dt = start + datetime.timedelta(days=_i)
    dt_list.append(dt)

# データ取得日リストに従いTwitterからデータ取得
for dt in dt_list:
    _start = dt.strftime('%Y-%m-%d')
    _end   = (dt + datetime.timedelta(days=1)).strftime('%Y-%m-%d')
    print('running ...', _start)

    # twintコマンド作成
    save_file = os.path.join(out_dir, f'tweet_uma_{dt.strftime('%Y%m%d')}.json')
    cmd = ['twint', 
           '--search', 'ウマ娘', 
           '--since', _start, 
           '--until', _end, 
           '--json', '--output', save_file]
    
    # twintでデータ取得
    res = subprocess.run(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
```
